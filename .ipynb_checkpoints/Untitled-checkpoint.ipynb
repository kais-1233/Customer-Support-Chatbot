```python
# step 2: Prepare Training Data  
# Extract all patterns and their intent tags from `intents.json` to create a clean DataFrame for model training and analysis.
 

patterns = []
tags = []

for intent in intents['intents']:
    tag = intent['tag']
    for pattern in intent['patterns']:
        patterns.append(pattern)
        tags.append(tag)

df = pd.DataFrame({'pattern': patterns, 'tag': tags})
print(f"Total training samples: {len(df)} | Unique intents: {df['tag'].nunique()}")
```

```python
# Step 3: Text Normalization  
# Clean and standardize all patterns using `TextBlob` for spelling correction and lowercase conversion to ensure consistent text input for model training.

def normalize_text(text):
    return text.strip().lower()

# Apply normalization
df['clean'] = df['pattern'].apply(lambda x: str(TextBlob(x).correct()).lower())
```

```python
# Step 4: Feature Extraction & Label Encoding
# Convert cleaned text to TF-IDF features (`X`).
# Encode intent labels (`y`) using `LabelEncoder`.

vectorizer = TfidfVectorizer(ngram_range=(1,2), min_df=1)
X = vectorizer.fit_transform(df['clean'])

le = LabelEncoder()
y = le.fit_transform(df['tag'])
```

```python
# Step 5: Model Training

model = LogisticRegression(max_iter=1000, multi_class='auto')
model.fit(X, y)
```

```python
# step 6: Evaluate on training set

y_pred = model.predict(X)
acc = accuracy_score(y, y_pred)
print(f"\nModel accuracy on training data: {acc*100:.2f}%")
print("\nClassification report:")
print(classification_report(y, y_pred, target_names=le.classes_))

```

```python
# step:7 Intent distribution plot
plt.figure(figsize=(12,4))
order = df['tag'].value_counts().index
sns.countplot(x='tag', data=df, order=order)
plt.xticks(rotation=60)
plt.title("Intent patterns distribution")
plt.tight_layout()
plt.show()
```
